```python
import tensorflow as tf

from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
import numpy as np
import glob, re, sys, random, time
from collections import deque

np.set_printoptions(threshold=sys.maxsize)
```

```python
THRESHOLD = 0.8

samples_raw = np.load("dataset/full_numpy_bitmap_cat.npy")[:50000]
thresholded = samples_raw / 255.0
thresholded[thresholded > THRESHOLD] = 1
thresholded[thresholded <= THRESHOLD] = 0

samples = samples_raw.astype(int )
```

```python
NUM_ACTIONS = 5

model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 2)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.Dropout(0.4))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(NUM_ACTIONS))
model.compile(loss="mean_squared_error", 
    optimizer="adam")
```

```python
MAX_TRIALS = 20000
THRESHOLD = 0.8
BATCH_SIZE = 64

def find_nearest_pos(pos, target, canvas):
    min_dist = 1000000000
    near = None
    
    remaining = target - canvas
    for x in range(remaining.shape[0]):
        for y in range(remaining.shape[1]):
            if remaining[x,y] == 1:
                dist = np.sqrt((x-pos[0])**2 + (y-pos[1])**2)
                if dist < min_dist:
                    near = np.array([x,y])
                    min_dist = dist
    
    #print(min_dist)
    return near

memory = deque(maxlen=50000)

for trial in range(MAX_TRIALS):
    print("starting trial", trial)
    target = samples[np.random.randint(0, samples.shape[0])].reshape(28,28)
    target = target.copy()
    target[target > THRESHOLD] = 1
    target[target <= THRESHOLD] = 0
    
    canvas = np.zeros((28, 28))
    pos = np.random.randint(0,28,size=(2))
    
    # While not done
    while np.sum(target - canvas) > 4:
        nearest_pos = find_nearest_pos(pos, target, canvas)
        
        correct_action = -1
        if np.all(pos == nearest_pos):
            correct_action = 4
        else:
            if np.abs(pos[0] - nearest_pos[0]) > np.abs(pos[1] - nearest_pos[1]):
                if pos[0] >= nearest_pos[0]:
                    correct_action = 3
                else:
                    correct_action = 2
            else:
                if pos[1] >= nearest_pos[1]:
                    correct_action = 1
                else:
                    correct_action = 0
        
        if correct_action >= 0:
            brush_map = np.zeros((28, 28))
            brush_map[pos[0], pos[1]] = 1
            
            label = np.zeros(NUM_ACTIONS)
            label[correct_action] = 1
        
            state = np.dstack([canvas.copy(), brush_map])
            
            memory.append([state, label])
            
            """
            output = target.copy()
            output[pos[0], pos[1]] = 2
            output[nearest_pos[0], nearest_pos[1]] = 3
            plt.imshow(output, cmap="gray")
            plt.show()
            """
            
            if correct_action == 0:
                if pos[1] < 27:
                    pos = pos + [0, 1]
            elif correct_action == 1:
                if pos[1] > 0:
                    pos = pos - [0, 1]
            elif correct_action == 2:
                if pos[0] < 27:
                    pos = pos + [1, 0]
            elif correct_action == 3:
                if pos[0] > 0:
                    pos = pos - [1, 0]
            elif correct_action == 4:
                canvas[pos[0], pos[1]] = 1
    
    if len(memory) > BATCH_SIZE:
        batch = random.sample(memory, BATCH_SIZE)
        images = []
        labels = []
        for image, label in batch:
            images.append(image)
            labels.append(label)
        
        images = np.array(images)
        labels = np.array(labels)
        model.fit(images, labels, epochs=1, verbose=False)
        
    #break
        
            
        
    
```
